```python
import requests
from bs4 import BeautifulSoup
```

```python
#url of web page 1
r = requests.get("https://www.century21.com/real-estate/chicago-il/LCILCHICAGO/?beds=1&baths=1&minsqft=200")

#c will now contain the HTML content of the web page at the given URL, which can then be parsed or processed for further use
c = r.content 

c
```

```python
soup=BeautifulSoup(c, "html.parser")
print(type(soup))
print(soup.prettify())
#now you have the whole html for this page
```

```python
#now we will go to the page , then detect the large container , we will notice it is  a div and its name is main-listiung row........
total = soup.find_all("div", {"class":"main-listings row row-cols-1 row-cols-md-2 row-cols-lg-3 row-cols-xxl-4 g-3 px-3 pe-sm-0 pb-3"})
#notice total now it is list have the div that have this class (actully the code have only one)
print(type(total))
print(len(total))
```

```python
# total of [0] mean i know the first elemnt in the list (the only div it have) , now i wanna know the beds , so i will do same thing in (inspect) to know what is the classs of beds
beds=total[0].find_all("div",{"class":"property-spec beds"})
len(beds) # same to the previous , it will be in beds variable as list homes have num of beds of them , print beds to notice that
print(beds)
```

```python
beds[0] 
```

```python
beds[0].text # will extract the string in html code 
```

```python
beds[0].text.split()
```

```python
int(beds[0].text.split()[0])
```

```python
for i in range(0,len(beds)):
    beds[i]=int(beds[i].text.split()[0])
beds
```

```python
# the same thing to prices
prices=total[0].find_all("div",{"class":"font-family-taglines property-price"})
len(prices)
```

```python
prices[0].text.strip().replace(",","")
```

```python
int(prices[0].text.strip().replace(",","").replace("$",""))
```

```python
for i in range(0,len(prices)):
    prices[i]=int(prices[i].text.strip().replace(",","").replace("$",""))
prices
```

### complete the same for rest properties :  number of bath , square feet and address

```python
# the same thing to square feet
sqft=total[0].find_all("div",{"class":"property-spec square-footage"})
len(sqft)
sqft
```

```python
int(sqft[0].text.split()[0].replace(",",""))
```

```python
for i in range(0,len(sqft)):
    sqft[i]=int(sqft[i].text.split()[0].replace(",",""))
sqft
```

```python
# the same thing to fall bathes
bath=total[0].find_all("span",{"class":"specs-number full-baths"})
len(bath)
bath
```

```python
int(bath[0].text)
```

```python
for i in range(0,len(bath)):
    bath[i]=int(bath[i].text)
bath
```

```python
# the same thing to address 
adrs=total[0].find_all("p",{"class":"property-address"})
len(adrs)
adrs
```

```python
adrs[0].text
```

```python
# now we cannot use strip() becuase it will delete only the \n in the sides without the middle , so use replace()
adrs[0].text.replace("\n","")
```

```python
for i in range(0,len(adrs)):
    adrs[i]=adrs[i].text.replace("\n","")
adrs
```
